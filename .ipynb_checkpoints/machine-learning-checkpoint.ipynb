## 1. Introduction of the Dataset

### Attributes

1. **Gender**: student's gender 
2. **NationalITy**: student's nationality 
3. **PlaceofBirth**: student's Place of birth 
4. **StageID**: educational level student belongs 
5. **GradeID**: grade student belongs 

6. **SectionID**: classroom student belongs 

7. **Topic**: Majoring in case of different courses
    
8. **Semester**: Number of semesters at school

9. **Relation**: Which parent is mainly responsible for students' daily life

10. **raisedhands**: how many times the student raises his/her hand on classroom 

11. **VisITedResources**: how many times the student visits a course content

12. **AnnouncementsView**: how many times the student checks the new announcements

13. **Discussion**: how many times the student participate on discussion groups 

14. **ParentAnsweringSurvey**: parent answered the surveys which are provided from school or not 

15. **ParentschoolSatisfaction**: the Degree of parent satisfaction from school

16. **StudentAbsenceDays**: the number of absence days for each student 

## 2.  Data Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

#Load the dataset
data=pd.read_csv('/Users/mac/Documents/2/Learning Analytics/SVM/students_data.csv')
data.head() #Display the first 5 columns of data
```

```python
#Check that the various fields with missing values

print(data.isnull().sum())
```

```python
#Check the feature type of each attribute

print(data.dtypes)
```

```python
#Data preprocessing

x=data.drop('Class',axis=1)
y=data['Class']
x=pd.get_dummies(x)#All the non-numeric variables into numeric
x.head()
```

```python
from sklearn.model_selection import train_test_split

#Select training and test sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)

```

# 3.  Predictive analysis of student performance

 ## 3-1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

#Use Logistic Regression model prediction and test accuracy

lr=LogisticRegression()
lr.fit(x_train,y_train)

predict_y=lr.predict(x_test)
print('predict',predict_y)

score=accuracy_score(y_test,predict_y)
print('Logistic Regression Accuracy：', score)
```

## - 3-2. NEW Logistic Regression（Remove unimportant variables)

```python
#Remove not important feature, Use logistic regression model again to compare results

df=data
x=df.drop('Class',axis=1)
x=x.drop(['StageID', 'Semester','SectionID'],axis=1)#remove 'StageID','Semester','SectionID'

y=df['Class']
x=pd.get_dummies(x)
x.head(10)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)

#Train the Logistic Regression model and test accuracy
lr=LogisticRegression()
lr.fit(x_train,y_train)

predict_y =lr.predict(x_test)
print('predict',predict_y)

score=accuracy_score(y_test,predict_y)
print('NEW Logistic Regression Accuracy：', score)
```

## - 3-3. SVM

```python
from sklearn import svm

#Train the SVM model and test accuracy
model = svm.SVC(kernel='linear', C=12)
model.fit(x_train,y_train)
model.score(x_test,y_test)

predicted = model.predict(x_test)
score1 =accuracy_score(y_test,predicted)
print(predicted)
print('SVM Precision Accuracy',score1)
```

## - 3-4. Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier

#Random Forest
rfc_model = RandomForestClassifier()
rfc_model.fit(x_train, y_train)

predicted = rfc_model.predict(x_test)
score1 =accuracy_score(y_test,predicted)
print(predicted)
print('Random Forest Precision Accuracy',score1)
```

## - 3-5. KNN

```python
from sklearn.neighbors import KNeighborsClassifier
#K-NN
model = KNeighborsClassifier(n_neighbors=7)
model.fit(x_train, y_train)#建立模型

predict_y = model.predict(x_test)
print('predict',predict_y)
score2 =accuracy_score(y_test,predict_y)
print('K-NN Precision Accuracy：', score2)

```
